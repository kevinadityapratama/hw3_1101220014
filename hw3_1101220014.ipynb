```python
import pandas as pd
```

Impor data dalam format txt

```python
zh = pd.read_csv("LB_TA.txt", delimiter=r"\s+")
zh
```

 Menghapus baris yang memiliki nilai "at" yang hilang

```python
zh_clean = zh.dropna(subset=['at'])
zh_clean
```

Mengelompokkan berdasarkan 'tic' dan menghitung rata-rata 'at' untuk setiap kelompok

```python
avg_at = zh_clean.groupby('tic')['at'].mean().reset_index()
avg_at
```

Menghapus 'tic' yang hanya memiliki satu record (dengan nilai 'at' yang tidak hilang)

```python
tic_counts = zh_clean['tic'].value_counts()
valid_tics = tic_counts[tic_counts > 1].index
avg_at = avg_at[avg_at['tic'].isin(valid_tics)]
```

Menyimpan hasilnya ke dalam file output.csv

```python
avg_at.to_csv('hasil lb_ta.csv', index=False)
```
