```python
import pandas as pd
```

```python
zh = pd.read_excel("Customer Call List.xlsx")
zh
```

```python
zh = zh.drop_duplicates()
zh
```

```python
zh = zh.drop(columns = "Not_Useful_Column")
zh
```

Menghapus semua karakter 1, 2, 3, ., _, dan / dari sisi kiri dan kanan string dalam kolom "Last_Name"

```python
zh["Last_Name"] = zh["Last_Name"].str.strip("123._/")
zh
```

Hapus nan dan na di kolom PhoneNumber

```python
#zh["Phone_Number"] = zh["Phone_Number"].str.replace(r'[^0-9]', '', regex=True)
#zh["Phone_Number"] = zh["Phone_Number"].apply(lambda x: str(x))
#zh["Phone_Number"] = zh["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
zh["Phone_Number"] = zh["Phone_Number"].str.replace('nan--','')

zh["Phone_Number"] = zh["Phone_Number"].str.replace('--','')
zh
```

Memisahkan kolom address menjadi 3 bagian yaitu Street_Address, State, Zip_Code

```python
zh[["Street_Address", "State", "Zip_Code"]] = zh["Address"].str.split(',', n=2, expand=True)
zh
```

Mengubah Yes menjadi Y dan No menjadi N pada kolom Paying Customer

```python
zh["Paying Customer"] = zh["Paying Customer"].str.replace('Yes','Y')
zh["Paying Customer"] = zh["Paying Customer"].str.replace('No','N')
zh
```

Mengubah Yes menjadi Y dan No menjadi N pada kolom Do_Not_Contact

```python
zh["Do_Not_Contact"] = zh["Do_Not_Contact"].str.replace('Yes','Y')
zh["Do_Not_Contact"] = zh["Do_Not_Contact"].str.replace('No','N')
zh
```

Mengganti N/a dan NaN dengan string kosong

```python
#zh = zh.replace('N/a','')
#zh = zh.replace('NaN','')

zh = zh.fillna('')
zh
```

Menghapus data yang Do_Not_Contact bernilai Y

```python
for x in zh.index:
    if zh.loc[x, "Do_Not_Contact"] == 'Y':
        zh.drop(x, inplace=True)

zh
```

Menghapus data dengan Phone_Number yang kosong

```python
for x in zh.index:
    if zh.loc[x, "Phone_Number"] == '':
        zh.drop(x, inplace=True)

zh
```

Mengatur ulang indeks seperti dari awal

```python
zh = zh.reset_index(drop=True)
zh
```
